# 1. Array Thinking vs. Iteration
Traditional iterative approach uses explicit Python loops. It is slow because each iteration involves Python overhead.

Array thinking treats data as whole arrays and applies operations directly, leveraging fast compiled code.

```python
# Traditional Python loop
def square_python(numbers):
    result = []
    for x in numbers:
        result.append(x ** 2)
    return result

# NumPy array thinking
def square_numpy(arr):
    return arr ** 2
```

```python
nums = list(range(11))
print(square_python(nums))
print(square_numpy(np.array(nums)))
```

# 2.Vectorization
Vectorization is the process of replacing explicit loops with array‑level operations. NumPy’s vectorized functions run in C, making them much faster.

```python
# Example: element-wise multiplication of two large arrays
size = 10_000_000
a = np.random.rand(size)
b = np.random.rand(size)

# Vectorized multiplication
%timeit c = a * b
```

# 3. Broadcasting
Broadcasting allows operations on arrays of different shapes without making copies of data.

```python
# Add a 1D array to each row of a 2D array
matrix = np.array([[1, 2, 3],
                   [4, 5, 6]])
row_bias = np.array([10, 20, 30])

result = matrix + row_bias  # broadcasting happens automatically
print(result)
```

# 4. Memory Efficiency
NumPy stores data in contiguous memory blocks, uses strides, and provides views instead of copies whenever possible.

```python
arr = np.arange(1_000_000)

# A slice returns a view (no data copy)
view = arr[::2]
print(view.base is arr)   # True, view shares memory with arr
```

```python
# Reshape usually returns a view (if possible)
reshaped = arr.reshape(500_000, 2)
print(reshaped.base is arr)   # True
```

```python
# Explicit copy if needed
copy = arr.copy()
print(copy.base is arr)
```

Strides allow efficient transposition without moving data:

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])
print(arr.strides)
```

```python
transposed = arr.T
print(transposed.strides)
```

# 5. Universal Functions (ufuncs)
Ufuncs are fast, element‑wise functions that support broadcasting, type casting, and optional output arrays.

```python
arr = np.array([1, 2, 3])

print(np.exp(arr))          # exponential
print(np.log(arr))          # natural log
print(np.greater(arr, 2))   # comparison ufunc


result = np.empty_like(arr)
np.multiply(arr, 10, out=result)   # writes directly into 'result'
print(result)
```

# 6. Performance Comparison: Python Loop vs. NumPy
We will create a list/array of 4 million numbers and square each element. We measure:

```python
import time
import tracemalloc

# Number of elements
N = 4_000_000

# ---- Python list approach ----
print("Python list approach:")
data_py = list(range(N))

tracemalloc.start()
start_time = time.perf_counter()

squared_py = [x**2 for x in data_py]

end_time = time.perf_counter()
current, peak = tracemalloc.get_traced_memory()
tracemalloc.stop()

print(f"  Time: {end_time - start_time:.4f} seconds")
print(f"  Peak memory: {peak / 1024**2:.2f} MB")
print(f"  Result size: {len(squared_py)} elements")
```

```python
# ---- NumPy array approach ----
print("\nNumPy array approach:")
data_np = np.arange(N, dtype=np.int64)   # use int64 to match Python int size

tracemalloc.start()
start_time = time.perf_counter()

squared_np = data_np ** 2

end_time = time.perf_counter()
current, peak = tracemalloc.get_traced_memory()
tracemalloc.stop()

print(f"  Time: {end_time - start_time:.4f} seconds")
print(f"  Peak memory: {peak / 1024**2:.2f} MB")
print(f"  Result shape: {squared_np.shape}")
```
